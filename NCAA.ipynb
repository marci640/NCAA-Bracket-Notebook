```python
import pandas as pd
```

```python
bracket = pd.read_csv('ncaa_bracket.csv', header=None, usecols=[2,4,7,10,13,14,17,22,26,29,32,35,37])
```

```python
bracket.rename(columns=
{
  2:"first_round",
  4: "second_round", 
  7: "regional_semifinals", 
 10: "regional_finals", 
 13: "final_four", 
 14: "national_championship",
 17: "winner",
 22: "national_championship_2",
 26: "final_four_2", 
 29: "regional_finals_2", 
 32: "regional_semifinals_2", 
 35: "second_round_2", 
 37: "first_round_2" 
},
inplace=True)
```

```python
import requests
from requests.auth import HTTPDigestAuth
import json
```

```python
url = "https://mighty-mesa-91143.herokuapp.com/api/v1/conferences"
myResponse = requests.get(url, verify=True)
bracket_data = []
if(myResponse.ok):
    # Loading the response data into a dict variable
    # json.loads takes in only binary or string variables so using content to fetch binary content
    # Loads (Load String) takes a Json file and converts into python data structure (dict or list, depending on JSON)
    jData = json.loads(myResponse.content)

    for key in jData:
        bracket_data.append(key)
else:
  # If response code is not ok (200), print the resulting http error code with description
    myResponse.raise_for_status()
```

```python
second_round = bracket['second_round'].append(bracket['second_round_2']).drop_duplicates()
regional_semifinals = bracket['regional_semifinals'].append(bracket['regional_semifinals_2']).drop_duplicates()
regional_finals = bracket['regional_finals'].append(bracket['regional_finals_2']).drop_duplicates()
final_four = bracket['final_four'].append(bracket['final_four_2']).drop_duplicates()
national_championship = bracket['national_championship'].append(bracket['national_championship_2']).drop_duplicates()
```

```python
bracket_data
```

```python
points = 0
for conference in bracket_data:
    for team in conference['teams']:
        if team['second_round'] and second_round.str.contains(team['college_name']).any():
            points += 1
        if team['regional_semifinals'] and regional_semifinals.str.contains(team['college_name']).any():
            points += 2
        if team['regional_finals'] and regional_finals.str.contains(team['college_name']).any():
            points += 4
        if team['final_four'] and final_four.str.contains(team['college_name']).any():
            points += 8
        if team['national_championship'] and national_championship.str.contains(team['college_name']).any():
            points += 16
        if team['winner'] and bracket['winner'].str.contains(team['college_name']).any():
            points += 32    
```
